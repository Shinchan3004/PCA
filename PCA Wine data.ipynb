```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine = pd.read_csv("C:/Users/SHIVAM/OneDrive/Desktop/Data science Assignment/PCA/wine.csv")
```

```python
wine
```

```python
wine['Type'].value_counts()
```

```python
Class = wine.iloc[:,0]
df = wine.iloc[:,1:]
```

```python
df.isnull().sum()
```

```python
df_norm = scale(df)
df_norm
```

```python
pca = PCA()
pca_values = pca.fit_transform(df_norm)
```

```python
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color="red")
```

```python
final_df = pd.DataFrame(pca_values[:,0:8],columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8'])
```

```python
final_df['Type'] = Class
final_df
```

## Heirarchial Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
df_std = final_df.iloc[:,0:3]
df_std
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_std, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=4,affinity= 'euclidean', linkage = 'single')
```

```python
y_hc = hc.fit_predict(df_std)
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters.value_counts()
```

```python
final_df['HC Clusters'] = Clusters
final_df
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df_std)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters = KMeans(3,random_state=40)
clusters.fit(df_std)
```

```python
clusters.labels_
```

```python
final_df['KMeans Clusters'] = clusters.labels_
```
